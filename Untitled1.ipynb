```python
import pandas as pd
```

```python
br_df = pd.read_csv('BR.csv', header=None)
uk_df = pd.read_csv('UK.csv', header=None)

def proc(df):
    df2 = df
    df3 = df
    df3 = df2.groupby(by=df2.loc[:,0]//24)
    df3=df3.sum()
    df3.loc[:,0] = (df3.loc[:,0] - 276)//576
    df3.loc[:,1] = df3.loc[:,1].divide(df3.loc[:,2], axis=0)
    df3.loc[:,3:] = df3.loc[:,3:].divide(df3.loc[:,2], axis=0)
    return df3
br_df.loc[:,0] = (br_df.loc[:,0] + 164) % 168
br_df = proc(br_df)
uk_df = proc(uk_df)
br_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1, 1, figsize=(5, 3))

titles=['acousticness', 'count', 'danceability', 'energy', 'instrumentalness', 'liveness',
          'loudness', 'speechiness', 'tempo', 'valence']

i = 10
br = br_df.loc[:, i]
uk = uk_df.loc[:, i]
x = i - 1
axs.plot(br_df.loc[:, 0], br, 'g--')
axs.plot(uk_df.loc[:, 0], uk, 'b-')
axs.set(title=titles[x])
plt.savefig('imgs/'+titles[i-1]+'_weekly.png')



plt.show()
```

## Analise
Os dados foram de 100 usuarios de cada pais, então ha ruido, no entanto algumas observaçoes podem ser feitas. o eixo x mostra quantas horas se passaram de meia noite da ultima segunda em relação a quando a musica foi ouvida. Outra observacão sobre os dados é o fato de que são dados recentes, então a estação do ano influencia nos horarios(em londres o dia comeca mais cedo 4 AM no verão).

### Contagem
Como se pode observar na grafico de contagem, as pessoas começam a escutar musicas mais cedo em londres, dado o grafico e esse grafico deixa claro a variação entre dia e noite.

### Danceability
As musicas do UK sao mais aptas para se dançar(de acordo com os classifcadores do netflix), praticamente em todos os horarios, o que indica que contra expectativas o UK curte ouvir mais musicas "Dançantes" que o BR no spotify.
###### Descricao do spotify: Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

```python
print("BR: ", br_df.iloc[:,3].sum() / br_df.iloc[:,3].count())
print("UK: ", uk_df.iloc[:,3].sum() / uk_df.iloc[:,3].count())
```

### Acousticness
Ambos os paises, escutam musicas parecidas no sentido de quão acusticas são. No entanto durante o periodo noturno as musicas do UK ficam muito mais acusticas(A confianca media que uma musica é acustica praticamente dobra nesse periodo), e uma possivel explicação para isso é que musicas acusticas são em geral mais lentas e calmas, então são melhores para dormir.
###### Descrição: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.

### Loudness & Tempo
Indicam o volume e a velocidade(em batidas por minuto) das musicas, e a informação obtida com a acousticness parece estar correta ao checar os dados, as subidas em acousticness, concordam com as decidas em tempo e loudness, o que é um indicativo que a explicação esteja correta.

### energy
Mais uma vez as musicas do UK parecem ser mais animadas que as brasileiras. E fora um pico pela manha, quanto mais tarde fica, mais aparenta energico.
#### Checando a energia no começo das noites para saber se o estilo anima perto da noite e em quais dias

```python
fig, axs = plt.subplots(7, 1, figsize=(10, 42))

titles=['SEG','TER','QUA','QUI','SEX','SAB','DOM']
for x in range(0,7):
    br = (br_df.loc[:,0]<(24+24*x)) & (br_df.loc[:,0]>=(24*x))
    #print(br)
    axs[x].plot(br_df.loc[br, 0]%24, br_df.loc[br, 3], color='green')
    axs[x].plot(uk_df.loc[br, 0]%24, uk_df.loc[br, 3], color='blue')
    axs[x].set(title=titles[x])



plt.show()
```
