```python
import pandas as pd
```

```python
br_df = pd.read_csv('Brazil.csv', header=None)
uk_df = pd.read_csv('UK.csv', header=None)

def proc(df):
    df2 = df.divide(df.loc[:,2],axis=0)
    df2.loc[:,2] = df.loc[:,2]
    df2.loc[:,0] = df.loc[:,0]
    return df2

br_df = proc(br_df)
uk_df = proc(uk_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(5, 2, figsize=(20, 40))

titles=['acousticness', 'count', 'danceability', 'energy', 'instrumentalness', 'liveness',
          'loudness', 'speechiness', 'tempo', 'valence']
for i in range(1,11):
    br = br_df.loc[:, i]
    uk = uk_df.loc[:, i]
    x = i - 1
    axs[x%5, x//5].plot(br_df.loc[:, 0], br, color='green')
    axs[x%5, x//5].plot(uk_df.loc[:, 0], uk, color='blue')
    axs[x%5, x//5].set(title=titles[x])



plt.show()
```

## Analise
Os dados foram de 100 usuarios de cada pais, ento ha ruido, no entanto algumas observaoes podem ser feitas. o eixo x mostra quantas horas se passaram de meia noite da ultima segunda em relao a quando a musica foi ouvida. Outra observaco sobre os dados  o fato de que so dados recentes, ento a estao do ano influencia nos horarios(em londres o dia comeca mais cedo 4 AM no vero).

### Contagem
Como se pode observar na grafico de contagem, as pessoas comeam a escutar musicas mais cedo em londres, dado o grafico e esse grafico deixa claro a variao entre dia e noite.

### Danceability
As musicas do UK sao mais aptas para se danar(de acordo com os classifcadores do netflix), praticamente em todos os horarios, o que indica que contra expectativas o UK curte ouvir mais musicas "Danantes" que o BR no spotify.
###### Descricao do spotify: Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

```python
print("BR: ", br_df.iloc[:,3].sum() / br_df.iloc[:,3].count())
print("UK: ", uk_df.iloc[:,3].sum() / uk_df.iloc[:,3].count())
```

### Acousticness
Ambos os paises, escutam musicas parecidas no sentido de quo acusticas so. No entanto durante o periodo noturno as musicas do UK ficam muito mais acusticas(A confianca media que uma musica  acustica praticamente dobra nesse periodo), e uma possivel explicao para isso  que musicas acusticas so em geral mais lentas e calmas, ento so melhores para dormir.
###### Descrio: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.

### Loudness & Tempo
Indicam o volume e a velocidade(em batidas por minuto) das musicas, e a informao obtida com a acousticness parece estar correta ao checar os dados, as subidas em acousticness, concordam com as decidas em tempo e loudness, o que  um indicativo que a explicao esteja correta.

### energy
Mais uma vez as musicas do UK parecem ser mais animadas que as brasileiras. 
#### Checando a energia no comeo das noites para saber se o estilo anima perto da noite e em quais dias

```python
fig, axs = plt.subplots(7, 1, figsize=(10, 42))

titles=['SEG','TER','QUA','QUI','SEX','SAB','DOM']
for x in range(0,7):
    br = (br_df.loc[:,0]<(24+24*x)) & (br_df.loc[:,0]>=(24*x))
    #print(br)
    axs[x].plot(br_df.loc[br, 0]%24, br_df.loc[br, 3], color='green')
    axs[x].plot(uk_df.loc[br, 0]%24, uk_df.loc[br, 3], color='blue')
    axs[x].set(title=titles[x])



plt.show()
```
