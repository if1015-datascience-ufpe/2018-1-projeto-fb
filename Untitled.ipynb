```python
import json, pylast, pprint, spotipy
from datetime import datetime 
pp = pprint.PrettyPrinter(indent=2)
from queue import Queue
file = open('cred.json', 'r')
creds = json.load(file)
api = creds['key']
secret = creds['secret']
lastfm = pylast.LastFMNetwork(api_key=api, api_secret=secret)

from spotipy import oauth2
sp_oauth = oauth2.SpotifyOAuth(client_id=creds["spot_id"],client_secret=creds["spot_secret"],redirect_uri="https://lalala/",scope="")
token_info = sp_oauth.get_cached_token() 
if not token_info:
    auth_url = sp_oauth.get_authorize_url()
    print(auth_url)
    response = input('Paste the above link into your browser, then paste the redirect url here: ')

    code = sp_oauth.parse_response_code(response)
    token_info = sp_oauth.get_access_token(code)

    token = token_info['access_token']

spotify = spotipy.Spotify(auth=token)
```

```python
import time
def is_token_expired(token_info):
    now = int(time.time())
    return token_info['expires_at'] - now < 60

def refresh():
    global token_info, spotify, sp_oauth
    if is_token_expired(token_info):
        token_info = sp_oauth.refresh_access_token(token_info['refresh_token'])
        token = token_info['access_token']
        spotify = spotipy.Spotify(auth=token)
        return spotify
```

```python
class TrackAggregation:
    def __init__(self, base=None):            
        self.acousticness = 0
        self.count = 0
        self.danceability = 0
        self.energy = 0
        self.instrumentalness = 0
        self.liveness = 0
        self.loudness = 0
        self.speechiness = 0
        self.tempo = 0
        self.valence = 0
        if base is not None:
            self.acousticness = base["acousticness"]
            self.count = 1
            self.danceability = base["danceability"]
            self.energy = base["energy"]
            self.instrumentalness = base["instrumentalness"]
            self.liveness = base["liveness"]
            self.loudness = base["loudness"]
            self.speechiness = base["speechiness"]
            self.tempo = base["tempo"]
            self.valence = base["valence"]
            
    def __str__(self):
        ret = "count:" + str(self.count) + "\n"
        ret += "acousticness:" + str(self.acousticness / max(self.count, 1)) + "\n"
        ret += "danceability:" + str(self.danceability / max(self.count, 1)) + "\n"
        ret += "energy:" + str(self.energy / max(self.count, 1)) + "\n"
        ret += "instrumentalness:" + str(self.instrumentalness / max(self.count, 1)) + "\n"
        ret += "liveness:" + str(self.liveness / max(self.count, 1)) + "\n"
        ret += "loudness:" + str(self.loudness / max(self.count, 1)) + "\n"
        ret += "speechiness:" + str(self.speechiness / max(self.count, 1)) + "\n"
        ret += "tempo:" + str(self.tempo / max(self.count, 1)) + "\n"
        ret += "valence:" + str(self.valence / max(self.count, 1)) + "\n"
        return ret
        
    def merge(self, aux):
        self.acousticness =  self.acousticness +  aux.acousticness
        self.count =  self.count +  aux.count
        self.danceability =  self.danceability +  aux.danceability
        self.energy =  self.energy +  aux.energy
        self.instrumentalness =  self.instrumentalness +  aux.instrumentalness
        self.liveness =  self.liveness +  aux.liveness
        self.loudness =  self.loudness +  aux.loudness
        self.speechiness =  self.speechiness +  aux.speechiness
        self.tempo =  self.tempo +  aux.tempo      
        self.valence =  self.valence +  aux.valence
        
        
class TrackStatistics:
    def __init__(self, timezone):  
        self.tz = timezone
        self.entries = [0] * 168
        for ind in range(0, 168):
            self.entries[ind] = TrackAggregation()
        
    def __str__(self):
        ret = str(self.tz) + ":\n"
        for ind in range(0,168):
            entry = self.entries[ind]
            ret += "---------\n"
            ret += str(ind // 24) + " - " + str(ind % 24) + "\n"
            ret += entry.__str__()
        return ret
            
    def add(self, ts: int, item: TrackAggregation):
        dt = datetime.fromtimestamp(ts)
        hour = dt.weekday() * 24 + dt.hour
        self.entries[hour].merge(item)
        

class DataSpider:
    def __init__(self, network, spotify):
        self.spotify = spotify
        self.status_user = {}
        self.br_queue = Queue()
        self.uk_queue = Queue()
        self.bad_queue = Queue()
        self.country_count = {}
        self.network = network
        self.memo_tracks = {}
        self.processed_country_count = {
            'Brazil': 0,
            'United Kingdom': 0,
        }
        self.track_data = {
            'Brazil': TrackStatistics('Brazil'),
            'United Kingdom': TrackStatistics('United Kingdom'),
        }
        
    def __str__(self):
        ret = ""
        for key in self.track_data:
            ret += "KEY: " + key +"\n"
            ret += self.track_data[key].__str__() + "\n"
        return ret
        
    def add_to_queue(self, user_name):
        if user_name in self.status_user:
            return
        user = pylast.User(user_name, self.network)
        try:
            country = user.get_country().get_name()
        except:
            return
        if country not in self.country_count:
            self.country_count[country] = 0
        self.country_count[country] += 1
        if country == 'Brazil':
            self.br_queue.put(user)
        if country == 'United Kingdom':
            self.uk_queue.put(user)
        else:
            self.bad_queue.put(user)
        self.status_user[user_name] = country
        
        
    def user_net(self, user):
        friends = user.get_friends(limit = 200)
        for friend in friends: 
            self.add_to_queue(friend.get_name())
        
        
    def run(self, rounds = 1):
        while rounds > 0:
            if self.processed_country_count['Brazil'] <= self.processed_country_count['United Kingdom'] + 10 and not self.br_queue.empty():
                user = self.br_queue.get()
                self.process_user(user)
                self.user_net(user)
                self.processed_country_count['Brazil'] += 1
            elif not self.uk_queue.empty():
                user = self.uk_queue.get()
                self.process_user(user)
                self.user_net(user)
                self.processed_country_count['United Kingdom'] += 1
            elif not self.bad_queue.empty():
                user = self.bad_queue.get()
                self.user_net(user)
            else:
                return "lacking users"
            rounds -= 1
            print("Remaining rounds: " + str(rounds))
                
    def track_genre(self, track):
        title = track.get_title()
        if title not in self.memo_tracks:
            try:
                aux = self.spotify.search(q='track:' + title, type='track', limit=1)
            except:
                self.spotify=refresh()
                aux = self.spotify.search(q='track:' + title, type='track', limit=1)
            tracks = aux['tracks']['items']
            if len(tracks) > 0:
                aux = tracks[0]
                uri = aux['uri']
                try:
                    aux = self.spotify.audio_features([uri])[0]
                except:
                    self.spotify = refresh()
                    aux = None
                self.memo_tracks[title] = TrackAggregation(aux)
            else:
                self.memo_tracks[title] = TrackAggregation()
        return self.memo_tracks[title]
    
    def process_user(self, user):
        user_name = user.get_name()
        country = self.status_user[user_name]
        tracks = user.get_recent_tracks(limit=1000)
        for track in tracks:
            trend = self.track_genre(track.track)
            ts = int(track.timestamp)
            self.track_data[country].add(ts, trend)
        self.status_user[user_name] += " OK"
    
    def create_csv(self, suffix):
        for key in self.track_data:
            csvv = open(key + "_" +str(suffix) + '.csv', 'w+')
            for ind in range(0, 168):
                ret = str(ind) + ", "
                ret += str(self.track_data[key].entries[ind].acousticness) + ", "
                ret += str(self.track_data[key].entries[ind].count) + ", "
                ret += str(self.track_data[key].entries[ind].danceability) + ", "
                ret += str(self.track_data[key].entries[ind].energy) + ", "
                ret += str(self.track_data[key].entries[ind].instrumentalness) + ", "
                ret += str(self.track_data[key].entries[ind].liveness) + ", "
                ret += str(self.track_data[key].entries[ind].loudness) + ", "
                ret += str(self.track_data[key].entries[ind].speechiness) + ", "
                ret += str(self.track_data[key].entries[ind].tempo) + ", "
                ret += str(self.track_data[key].entries[ind].valence) + "\n"
                csvv.write(ret)
            csvv.close()


 
dss = DataSpider(lastfm, spotify)
user_names = ['arthurlpgc', 'cobralibre']
for name in user_names:
    dss.add_to_queue(name)
```

```python
for ind in range(101, 200):
    try:
        dss.run(1)
    except:
        refresh()
        dss.spotify=spotify
    dss.create_csv(ind)
    
    print(dss.processed_country_count)
```
